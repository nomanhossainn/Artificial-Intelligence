```python
import pandas as pd

dataFrame = pd.read_csv(dataPath)
testDataFrame = pd.read_csv(testDataPath)
```

```python
print(dataFrame)
print(testDataFrame)
```

```python
dataSet = dataFrame.to_numpy()
testDataSet = testDataFrame.to_numpy()
```

```python
dataSet[np.where(dataSet=='Yes')] = 1
dataSet[np.where(dataSet=='No')] = 0
dataSet[np.where(dataSet=='Available')] = 1
dataSet[np.where(dataSet=='Not Available')] = 1
dataSet[np.where(dataSet=='Affordable')] = 0
dataSet[np.where(dataSet=='Budget')] = 1
dataSet[np.where(dataSet=='Premium')] = 2
dataSet[np.where(dataSet=='Luxury')] = 3

print(dataSet)


testDataSet[np.where(testDataSet=='Yes')] = 1
testDataSet[np.where(testDataSet=='No')] = 0
testDataSet[np.where(testDataSet=='Available')] = 1
testDataSet[np.where(testDataSet=='Not Available')] = 1
testDataSet[np.where(testDataSet=='Affordable')] = 0
testDataSet[np.where(testDataSet=='Budget')] = 1
testDataSet[np.where(testDataSet=='Premium')] = 2
testDataSet[np.where(testDataSet=='Luxury')] = 3

print(testDataSet)
```

```python
def eucledianDistance(x1, x2):
    distance = 0
    for i in range(len(x1)):
        distance += (x1[i] - x2[i]) ** 2
    return distance ** 0.5

def predict(x_train, y_train, x_test, k=3):
    distances = []
    for i in range(len(x_train)):
        d = eucledianDistance(x_train[i], x_test)
        distances.append(d)

    sorted_distances = np.argsort(distances)[:k]

    poll = dict()
    poll[0] = 0  # Label 0
    poll[1] = 0  # Label 1
    poll[2] = 0  # Label 2
    poll[3] = 0  # Label 3

    for i in sorted_distances:
        label = y_train[i]
        poll[label] += 1

    return max(poll, key=poll.get)


x_train = dataSet[:, 0:4]
y_train = dataSet[:, 4]

x_test = testDataSet[:, 0:4]
y_test = testDataSet[:, 4]

correct_predictions = 0
correctly_predicted_rows = []

for i in range(len(x_test)):
    predicted_label = predict(x_train, y_train, x_test[i], k=3)
    actual_label = y_test[i]

    if predicted_label == actual_label:
        correct_predictions += 1
        correctly_predicted_rows.append(np.concatenate((x_test[i], [actual_label])))

print(f"Number of correctly predicted rows: {correct_predictions}")
print("Correctly predicted rows:")
for row in correctly_predicted_rows:
    print(row)
```
