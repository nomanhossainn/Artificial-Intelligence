```python
x1 = np.array(data['x1'])
x2 = np.array(data['x2'])
x3 = np.array(data['x3'])
y = np.array(data['y'])
```

```python
print(x1)
print(x2)
print(x3)
print(y)
```

```python
n = len(y)
X = np.column_stack((np.ones(n), x1, x2, x3))
```

```python
X_transpose = X.T
X_transpose_X = np.dot(X_transpose, X)
X_transpose_X_inv = np.linalg.inv(X_transpose_X)
X_transpose_y = np.dot(X_transpose, y)
```

```python
coefficients = np.dot(X_transpose_X_inv, X_transpose_y)
a0, a1, a2, a3 = coefficients
```

```python
print(f"Intercept (a0): {a0}")
print(f"Coefficient for x1 (a1): {a1}")
print(f"Coefficient for x2 (a2): {a2}")
print(f"Coefficient for x3 (a3): {a3}")
```
